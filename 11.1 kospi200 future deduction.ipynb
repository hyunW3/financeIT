```python
import pandas as pd
import numpy as np
```

```python
# pd.read_excel(엑셀파일명,시트명,index_col=칼럼명)
df = pd.read_excel('09.2 KOSPI200 F 201712.xlsx',sheet_name = 'Sheet1',index_col= '일자')
df.index = pd.to_datetime(df.index)
df
```

```python
def futures_price(S,r,d,T,t0): # 함수선언 입력값으로 (기초자산 가격, 이자율, 배당률, 최종거래일, 기준일 )
    t = np.busday_count(t0,T) / 252
    print(t)
    F = S*np.exp((r-d)*t) # F= Se^((r-d)t)
    return F # 선물 이론가 반환
```

```python
T = pd.to_datetime('2017-12-14') # 최종거래일
r = 0.0225 #이자율
d = 0.015 # 배당률
for n in (df.index) :
    t0 = n
    S = df.loc[n,'지수']
    F = futures_price (S ,r ,d, T, t0)
    f = '{:,.2f}'.format(F) # 이론가 숫자 포매팅
    df.loc[n,'이론가'] = f
```
